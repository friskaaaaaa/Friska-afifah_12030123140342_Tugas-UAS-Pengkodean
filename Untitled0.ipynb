```python
import pandas as pd
import matplotlib.pyplot as plt

# Membaca data dari CSV
kebijakan_fiskal = pd.DataFrame({
    'tahun': [2023, 2024, 2025],
    'tax_rate': [0.22, 0.22, 0.22], --
    'tax_holiday_awal': [2023, 2023, 2023],
    'tax_holiday_akhir': [2027, 2027, 2027]
})

aset_tetap = pd.DataFrame({
    'aset_id': [1, 2],
    'kategori': ['Mesin Jahit', 'Kendaraan Operasional'],
    'nilai_perolehan': [100000000, 150000000],
    'umur_ekonomis': [5, 4],
    'metode': ['garis_lurus', 'saldo_menurun']
})

transaksi_keuangan = pd.DataFrame({
    'tahun': [2023, 2023, 2024, 2024],
    'pendapatan': [500000000, 600000000, 700000000, 800000000],
    'beban_operasional': [200000000, 250000000, 300000000, 350000000],
    'penyusutan': [50000000, 60000000, 70000000, 80000000],
    'skenario': ['normal', 'tax_holiday', 'normal', 'tax_holiday']
})

# Fungsi untuk menghitung penyusutan
def hitung_penyusutan(nilai_perolehan, umur_ekonomis, metode, tahun):
    if metode == 'garis_lurus':
        return nilai_perolehan / umur_ekonomis
    elif metode == 'saldo_menurun':
        return (2 * nilai_perolehan) / umur_ekonomis  # Simplified double declining balance

# Menghitung penyusutan untuk setiap aset
penyusutan_per_aset = []
for _, aset in aset_tetap.iterrows():
    for tahun in [2023, 2024]:
        penyusutan = hitung_penyusutan(aset['nilai_perolehan'], aset['umur_ekonomis'], aset['metode'], tahun)
        penyusutan_per_aset.append({
            'tahun': tahun,
            'kategori': aset['kategori'],
            'penyusutan': penyusutan,
            'metode': aset['metode']
        })

df_penyusutan = pd.DataFrame(penyusutan_per_aset)

# Menghitung PPh Badan
def hitung_pph_badan(pendapatan, beban_operasional, penyusutan, tax_rate, skenario, tahun, tax_holiday_awal, tax_holiday_akhir):
    laba_sebelum_pajak = pendapatan - beban_operasional - penyusutan
    if skenario == 'tax_holiday' and tax_holiday_awal <= tahun <= tax_holiday_akhir:
        pph_badan = 0  # Tax holiday berarti PPh = 0
    else:
        pph_badan = laba_sebelum_pajak * tax_rate
    return laba_sebelum_pajak, pph_badan

# Menghitung PPh Badan untuk setiap skenario
result = []
for _, row in transaksi_keuangan.iterrows():
    tahun = row['tahun']
    tax_rate = kebijakan_fiskal[kebijakan_fiskal['tahun'] == tahun]['tax_rate'].iloc[0]
    tax_holiday_awal = kebijakan_fiskal[kebijakan_fiskal['tahun'] == tahun]['tax_holiday_awal'].iloc[0]
    tax_holiday_akhir = kebijakan_fiskal[kebijakan_fiskal['tahun'] == tahun]['tax_holiday_akhir'].iloc[0]

    laba, pph = hitung_pph_badan(
        row['pendapatan'], row['beban_operasional'], row['penyusutan'],
        tax_rate, row['skenario'], row['tahun'], tax_holiday_awal, tax_holiday_akhir
    )

    result.append({
        'tahun': row['tahun'],
        'skenario': row['skenario'],
        'pendapatan': row['pendapatan'],
        'beban_operasional': row['beban_operasional'],
        'penyusutan': row['penyusutan'],
        'laba_sebelum_pajak': laba,
        'pph_badan': pph
    })

df_result = pd.DataFrame(result)

# Menghitung penyusutan untuk skenario garis lurus dan saldo menurun
penyusutan_result = []
for _, aset in aset_tetap.iterrows():
    for tahun in [2023, 2024]:
        for metode in ['garis_lurus', 'saldo_menurun']:
            penyusutan = hitung_penyusutan(aset['nilai_perolehan'], aset['umur_ekonomis'], metode, tahun)
            penyusutan_result.append({
                'tahun': tahun,
                'kategori': aset['kategori'],
                'penyusutan': penyusutan,
                'metode': metode
            })

df_penyusutan_skenario = pd.DataFrame(penyusutan_result)

# Menampilkan hasil tabel
print("Hasil Perhitungan PPh Badan:")
print(df_result)

print("\nHasil Perhitungan Penyusutan (Garis Lurus dan Saldo Menurun):")
print(df_penyusutan_skenario)

# Visualisasi PPh Badan
plt.figure(figsize=(10, 6))
for skenario in df_result['skenario'].unique():
    data = df_result[df_result['skenario'] == skenario]
    plt.bar(data['tahun'].astype(str) + '_' + skenario, data['pph_badan'], label=skenario)

plt.title('PPh Badan per Tahun dan Skenario')
plt.xlabel('Tahun dan Skenario')
plt.ylabel('PPh Badan (IDR)')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Visualisasi Penyusutan
plt.figure(figsize=(10, 6))
for metode in df_penyusutan_skenario['metode'].unique():
    data = df_penyusutan_skenario[df_penyusutan_skenario['metode'] == metode]
    plt.bar(data['tahun'].astype(str) + '_' + data['kategori'] + '_' + metode,
            data['penyusutan'], label=metode)

plt.title('Penyusutan per Tahun, Kategori, dan Metode')
plt.xlabel('Tahun, Kategori, dan Metode')
plt.ylabel('Penyusutan (IDR)')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
